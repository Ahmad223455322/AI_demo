```python
sensor = 'sensor_data_2022119_1028.csv'
df_sensor = pd.read_csv(sensor, sep= ',', header = 0)
```

```python
df_sensor.head()
```

```python
del df_sensor['Unnamed: 0']
df_sensor.head()
```

```python
df_sensor.rename(columns = {'time':'Date'}, inplace = True)
```

```python
df_sensor.head()
```

```python
df_sensor[['Date','Time']] = df_sensor.Date.str.split(expand=True)
df_sensor.head()
```

```python
df_sensor['Date'] = df_sensor['Date'].map(lambda x: x.lstrip(',').rstrip(','))
df_sensor['Time'] = df_sensor['Time'].map(lambda x: x.lstrip(' ').rstrip(' '))
df_sensor = df_sensor.replace('/','-', regex=True)
df_sensor = df_sensor.replace(';',':', regex=True)
df_sensor.head()
```

```python
df_sensor["Sensor address"].values
```

```python
df_sensor.loc[df_sensor["Sensor address"] == "Madelens skrivbord Addiva Sigurdsgatan", "Sensor address"] = "1"
df_sensor.loc[df_sensor["Sensor address"] == "Skrivbord Addiva Sigurdsgatan", "Sensor address"] = "2"
df_sensor.loc[df_sensor["Sensor address"] == "Matsal Addiva Sigurdsgatan", "Sensor address"] = "3"
df_sensor.loc[df_sensor["Sensor address"] == "Pingisrum Addiva Sigurdsgatan", "Sensor address"] = "4"
df_sensor.loc[df_sensor["Sensor address"] == "ERS Sensor Addiva Sigurdsgatan", "Sensor address"] = "5"
```

```python
df_sensor.tail()
```

```python
df_sensor = df_sensor.rename({'Sensor address': 'sensor name',"Date":"date", "Time":"time"}, axis='columns')
```

```python
df_sensor.insert(0, 'time', df_sensor.pop('time'))
df_sensor.insert(0, 'date', df_sensor.pop('date'))
```

```python
df_sensor['hour'] = df_sensor['time']
df_sensor.tail()
```

```python
df_sensor["hour"] = df_sensor["hour"].str[:-6]
df_sensor.head()
```

```python
outdoor_temp = '20221109_smhi_data_vasteras.csv'
df_outdoor = pd.read_csv(outdoor_temp, sep= ';', header = 0)
df_outdoor.head()
```

```python
df_outdoor['hour'] = df_outdoor['time']

df_outdoor["hour"] = df_outdoor["hour"].str[:-6]
df_outdoor.head()
```
