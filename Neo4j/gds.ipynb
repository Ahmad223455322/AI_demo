### Projecting into GDS

```python
result = gds.run_cypher(
  """
    CALL gds.graph.project(
      'mygraph',
      ['*'],
      {
          relationshipProjection:'*'
        
      }
    )
    YIELD
      graphName AS graph,
      relationshipProjection AS readProjection,
      nodeCount AS nodes,
      relationshipCount AS rels
  """
)
display(result)
```

```python

result = gds.run_cypher(
  """
    CALL gds.graph.list()
  """
)
display(result)
```

```python
result = gds.run_cypher(
  """
  CALL gds.fastRP.mutate('mygraph',{
    embeddingDimension: 16,
    randomSeed: 1,
    mutateProperty:'embedding'
  })
  """
)
display(result)
```

```python
result = gds.run_cypher(
  """
    CALL gds.graph.writeNodeProperties('mygraph', ['embedding'])
    YIELD writeMillis
  """
)
display(result)
```

```python
result = gds.run_cypher(
  """
    MATCH (n:Sensor) RETURN n
  """
)
display(result)
```
